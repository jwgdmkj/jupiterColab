```python
from keras.utils import  np_utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense as D

# 전체 구현 ( batch_size =128)
# 첫번째 히든 레이어 유닛수를 1로 바꾼 결과 - 성능 하락 확인
# 세개의 히든 레이어 모두 삭제한 성능(14~16 line 삭제)
# E-mail : joongheon@korea.ac.kr
# [KIRO] name_

#####  함수 해석 #######

# 입력을 댄스라고 부름
# 숫자는 유닛의 갯수
# model.add =레이어 갯수 증가
# Keras는 레고 블럭처럼 코딩하는 방식을 추구 한다
# model.compile 연산방식 설정
# for문은 model.fit으로 사용
# model.evaluate 는 정확도 측정
# model.summary()는 시각화 테이블 생성
```

```python
# MNIST data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print(train_images.shape, train_labels.shape, test_images.shape, test_labels.shape)
train_images = train_images.reshape(train_images.shape[0],784).astype('float32')/255.0
test_images = test_images.reshape(test_images.shape[0],784).astype('float32')/255.0
train_labels = np_utils.to_categorical(train_labels) # One-Hot Encoding
test_labels = np_utils.to_categorical(test_labels) # One-Hot Encoding
```

```python
# Model
model = Sequential()
model.add(D(256, activation='relu'))   # units=256, activation="relu"
model.add(D(256, activation='relu'))   # units=256, activation="relu"
model.add(D(256, activation='relu'))   # units=256, activation="relu"
model.add(D(10, activation='softmax')) # units=10, activation="softmax"
model.compile(loss="categorical_crossentropy", optimizer="sgd",metrics=['accuracy'])
```

```python
# Training
model.fit(train_images, train_labels, epochs=5, verbose=1)
```

```python
# Testing
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()
```

```python
# 1. 전체 구현 (batch_size = 128)
# Training
model.fit(train_images, train_labels, epochs=5, batch_size=128, verbose=1)
# Testing
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()
```

```python
# 2. 첫번째 히든레이어 유닛 수를 1로 바꾼 결과
# Model
model = Sequential()
model.add(D(1, activation='relu'))   # units=1, activation="relu"
model.add(D(256, activation='relu'))   # units=256, activation="relu"
model.add(D(256, activation='relu'))   # units=256, activation="relu"
model.add(D(10, activation='softmax')) # units=10, activation="softmax"
model.compile(loss="categorical_crossentropy", optimizer="sgd",metrics=['accuracy'])

# Training
model.fit(train_images, train_labels, epochs=5, batch_size=32, verbose=1)
# Testing
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()

# Training2
model.fit(train_images, train_labels, epochs=5, batch_size=128, verbose=1)
# Testing2
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()
```

```python
# 3. 세개의 히든레이어 모두 삭제한 성능
# Model
model = Sequential()
model.add(D(10, activation='softmax')) # units=10, activation="softmax"
model.compile(loss="categorical_crossentropy", optimizer="sgd",metrics=['accuracy'])

# Training
model.fit(train_images, train_labels, epochs=5, batch_size=32, verbose=1)
# Testing
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()

# Training2
model.fit(train_images, train_labels, epochs=5, batch_size=128, verbose=1)
# Testing2
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()

# Training2
model.fit(train_images, train_labels, epochs=5, verbose=1)
# Testing2
cost, accuracy = model.evaluate(test_images, test_labels)
print("Accuracy: ", accuracy)
model.summary()
```
