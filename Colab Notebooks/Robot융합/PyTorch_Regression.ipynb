```python
import torch

x_data = torch.FloatTensor([[1,1], [2,2], [3,3]])
y_data = torch.FloatTensor([[10], [20], [30]])
W = torch.randn([2,1], requires_grad = True) #정규분포에서, 주어진 형태의 난수를 생성.
b = torch.randn([1], requires_grad = True) #즉, W는 [2,1]짜리 배열, b는 [1]짜리 데이터
optimizer = torch.optim.SGD([W,b], lr=0.01)
```

```python
#행렬 곱해, b 더하기
def model_LinearRegression(x):
  return torch.matmul(x, W) + b
```

```python
for step in range(2000):
  prediction = model_LinearRegression(x_data)
  cost = torch.mean((prediction - y_data) ** 2)

  optimizer.zero_grad() #역전파 단계 수행 실행 전, 변화도를 0으로
  # cost에, 어떤 스칼라 값에 대한 출력텐서의 변화도(gradient)를 전달받고,
  # 동일한 스칼라 값에 대한 입력텐서의 변화도를 계산
  cost.backward() 
  optimizer.step() 
```

```python
x_test = torch.FloatTensor([[4,4]])
model_test = model_LinearRegression(x_test) #임의의 4*4짜리 행렬 x_test에 계산
print("Model with [4] (expectation:40) : ", model_test.detach().item())
```
