```python
import torch
import torch.nn as nn
import torch.nn.functional as F

x_train = torch.FloatTensor([[1,1], [2,2], [3,3]])
y_train = torch.FloatTensor([[10], [20], [30]])
```

```python
class class_LinearRegression(nn.Module) :
  def __init__(self) :
    super().__init__()
    self.linear = nn.Linear(2,1)
  def forward(self, x) :
    return self.linear(x)
```

```python
model_LinearRegression = class_LinearRegression()
optimizer = torch.optim.SGD(model_LinearRegression.parameters(), lr = 0.01)
```

```python
for step in range(2000) :
  prediction = model_LinearRegression(x_train)
  cost = F.mse_loss(prediction, y_train)
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
```

```python
x_test = torch.FloatTensor([[4,4]])
model_test = model_LinearRegression(x_test)
print("Model with [4] (expectation:40) : ", model_test.detach().item())
```
